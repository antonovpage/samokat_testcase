```python
import pandas as pd
from datetime import datetime
```

```python
warehouses = pd.read_excel('warehouses.xlsx')
warehouses.head()
```

```python
products = pd.read_excel('products.xlsx')
products.head()
```

```python
orders = pd.read_excel('orders.xlsx')
orders.head()
```

```python
order_lines = pd.read_excel('order_lines.xlsx')
order_lines.head()
```

```python
products['category'].unique()
```

```python
products = products[products['category'] == 'Продукция для животных']
products = products[products['product'] != 'Корм Kitekat для кошек, с кроликом в соусе, 85 г']
products.head()
```

```python
#список айдишников кормов
keys = list(products['product_id'])
```

```python
#оставляем в таблице 'order_lines' только те заказы, которые включают корма для животных
order_lines = order_lines[order_lines.product_id.isin(keys)].reset_index()
del order_lines['price']
del order_lines['index']
order_lines
```

```python
order_lines.info()
```

```python
#ограничиваем столбец с датой заказа по 15.08.2017
orders = orders[orders['order_date'] <= '2017-08-15']
orders.head()
```

```python
keys_orders = list(orders['order_id'].unique())
keys_orders
```

```python
#фильтруем таблицу order_lines, оставляя только заказы, оформленные до 15.08.2017
order_lines = order_lines[order_lines.order_id.isin(keys_orders)]
order_lines.info()
```

```python
order_lines.head()
```

```python
order_lines = order_lines.merge(orders, how='left')
```

```python
order_lines
```

```python
del order_lines['warehouse_id']
del order_lines['order_date']
```

```python
order_lines
```

```python
users = order_lines.groupby('user_id')['product_id'].count().reset_index()
users.rename(columns = {'user_id' : 'user_id', 'product_id' : 'quantity'}, inplace = True)
users.head()
```

```python
users_final = users[users['quantity'] == 2]
users_final.reset_index(drop=True)
```

```python
users_final.to_csv("task2.csv")
```
